```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')

%config InlineBackend.figure_format = 'svg'
%matplotlib inline
```

```python
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
```

```python
boston.keys()
```

```python
data = boston.data
```

```python
feature_names = boston.feature_names
```

```python
X = pd.DataFrame(data, columns=feature_names)
```

```python
price = boston.target
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, price, test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
# создадим объект scaler:
scaler = StandardScaler()
```

```python
#  с помощью метода fit_transform построим модель масштабирования 
# и запишем результат в переменную X_train_scaled
X_train_scaled = scaler.fit_transform(X_train)
```

```python
from sklearn.manifold import TSNE
```

```python
tsne = TSNE(n_components=2, learning_rate=150, random_state=100)
```

```python
X_train_tsne = tsne.fit_transform(X_train_scaled)
```
